```python
#KNN
import pandas as pd
import numpy as np
from sklearn import neighbors, metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('car.data')
data.columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
X = data[['buying', 'maint', 'safety']].values
y = data[['class']]

Le = LabelEncoder()
for i in range(len(X[0])):
    X[:,i] = Le.fit_transform(X[:,i])


label_mapping = {'unacc':0, 'acc':1, 'good':2, 'vgood':3}
y['class'] = y['class'].map(label_mapping)
y = np.array(y)

a = []
lis = [i for i in range(1,100)]
for g in range(1,100):
    knn = neighbors.KNeighborsClassifier(n_neighbors=g, weights='uniform')
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_test)
    accuracy = metrics.accuracy_score(y_test, predictions)
    a.append(accuracy)
plt.plot(lis, a)

```

```python
#SVM
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

iris = datasets.load_iris()
#df = pd.DataFrame(iris)
#df.head()""""
X = iris.data 
y = iris.target

classes = ['Iris Setosa', 'Iris Versicolour', 'Iris Virginica']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = svm.SVC()
model.fit(X_train, y_train) #trains the model

prediction = model.predict(X_test)
accuracy = accuracy_score(y_test, prediction)

print(y_test)
print(prediction)
print(accuracy)
```

```python
#Linear Regression
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt

boston = datasets.load_boston()
X = boston.data
y = boston.target

l_reg = linear_model.LinearRegression()
for i in range(0,12):
    plt.scatter(X.T[i],y)
    plt.title(i+1)
    plt.show()
```

```python
#LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = l_reg.fit(X_train, y_train)
prediction = model.predict(X_test)

print(l_reg.score(X,y)) #R2 value
print(l_reg.coef_) #coefficeint value which is basically the slope of the line between the i-th feature and y-labels
print(l_reg.intercept_)
```

```python
#KMeans
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale
import pandas as pd

bc = datasets.load_breast_cancer()
X = scale(bc.data)
y = bc.target
X1 = pd.DataFrame(X)
X1.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = KMeans(n_clusters = 2, random_state = 0)
model.fit(X_train)
prediction = model.predict(X_test)
labels = model.labels_
for i in prediction:
    if i == 0:
        i = 1
    else:
        i = 0

print(labels)
print(prediction)
print(accuracy_score(y_test, prediction))
print(y_test)
print(prediction)
```

```python
#PCA
import sklearn
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
data1 = load_iris()
x = data1.data
y = data1.target
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_predict = lr.predict(x_test)
accuracy = accuracy_score(y_predict, y_test)
print(accuracy)
```
