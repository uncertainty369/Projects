## SUMMER SCHOOL TASK 1 - DATA VISUALISATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = "https://raw.githubusercontent.com/analytics-club-iitm/Summer-School-2022/main/Task_1/data/data.csv"
df = pd.read_csv(url)
df.head()
```

```python
df['Education'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(40, 30))
sns.countplot(x = "Remarks", data = df, ax = ax)
```

From this bar graph we can see that the majority of employees who left had an issue with the manager or felt like they their career was stagnant (lack of growth).

```python
df['Marital Status'].value_counts()
fig1, ax1 = plt.subplots()
ax1.pie([118, 106, 103], labels = ["Divorced", "Single", "Married"], autopct='%1.1f%%')
ax1.axis('equal')
```

From the pie chart we see that the employees who left are equally likely to be divorced, single or married.

```python
fig2, ax2 = plt.subplots(figsize=(10, 10))
sns.countplot(x = "Gender", data = df, ax = ax2)
```

An overwhelming percent of the employees who leave are male.

```python
df['count'] = 1
df_group_sal = df.groupby("Grade")[['Monthly Income']].mean()
df_group_sal_re = df_group_sal.reindex(['E1', 'E2', 'M1', 'M2', 'M3', 'M4', 'CXO'])
df_group = df.groupby("Grade")[['count']].sum()
df_group_re = df_group.reindex(['E1', 'E2', 'M1', 'M2', 'M3', 'M4', 'CXO'])
fig5, ax5 = plt.subplots(1,2, figsize = (20,6))
df_group_re.plot(kind = 'bar', ax = ax5[0], title = 'Employees in each Grade')
df_group_sal_re.plot(kind = 'bar', ax = ax5[1], color = 'r', title = 'Monthly Income (INR) vs Grade')
plt.show()
```

Most of the employees who leave are of a lower grade (lower pay).

```python
sns.set_style('whitegrid')
sns.displot(df['Age'], kde = False, color ='red', bins = 30)
```

Employees with the age in the range of 25-30 years are most probably to leave the company

```python
fig6, ax6 = plt.subplots(figsize=(10, 6))
sns.countplot(x = "Education", data = df, ax = ax6)
```

Most employees leave to pursue higher education as we can see this from the graph. Most of them who left have only done a Bachelors degree.
